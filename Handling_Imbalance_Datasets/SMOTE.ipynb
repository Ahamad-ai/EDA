# SMOTE

```python
from imblearn.over_sampling import SMOTE
import pandas as pd
```

```python
oversample=SMOTE()
```

```python
df=pd.read_csv('./data/raw/creditcard.csv')
```

```python
df["Class"].value_counts()
```

```python
X,y=oversample.fit_resample(df.drop("Class", axis=1), df["Class"])
```

```python
X.shape
```

```python
len(y[y==1])
```

```python
df1=pd.DataFrame(X, columns=df.columns[:-1])
df2=pd.DataFrame(y, columns=["Class"])

```

```python
oversample_df=pd.concat([df1, df2], axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(oversample_df[df1.columns[0]], oversample_df[df1.columns[1]], c=oversample_df["Class"], cmap='viridis', s=1)
```
